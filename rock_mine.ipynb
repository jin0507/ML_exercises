# Import Libraries

```python
import pandas as pd 
from sklearn.preprocessing import LabelEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
```

# Data

```python
data = pd.read_csv("D:/machine_learning/datasets/sonar_data.csv", header=None)
```

# Split train and test set

```python
x = data.drop([60], axis=1)
y = data[60]
```

```python
label_encode = LabelEncoder()
y = label_encode.fit_transform(y)
y[0]
```

* 0: Mine
* 1: Rock

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, shuffle=True, random_state=42)
```

# Use LogisticRegression as model

```python
log_reg = LogisticRegression()
log_reg.fit(x_train, y_train)
y_train_pred = log_reg.predict(x_train)
```

```python
y_train_pred
```

```python
accuracy_score(y_train, y_train_pred)
```

```python
results_test = pd.DataFrame(columns=["Classifier", "Accuracy"])
```

```python
import warnings
```

# Classifiers and Accuracy

```python
classifiers = {
    ("logistic", LogisticRegression()),
    ("tree", DecisionTreeClassifier()),
    ("neighbor", KNeighborsClassifier()),
    ("SVM", SVC())
}

with warnings.catch_warnings():
    warnings.filterwarnings("ignore", category=FutureWarning)
    for name, classifier in classifiers:
        model = classifier.fit(x_train, y_train)
        y_test_pred = model.predict(x_test)
        metrics_test = {
            "Classifier": name, 
            "Accuracy": accuracy_score(y_test, y_test_pred)
        }
        results_test = results_test.append(metrics_test, ignore_index=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_style("whitegrid")
sns.barplot(x="Classifier", y="Accuracy", data=results_test)
plt.title("Accuracy of Classifiers in Training set")
plt.show()
```
